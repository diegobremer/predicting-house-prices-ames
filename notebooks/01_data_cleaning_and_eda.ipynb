```python
#Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carregando a base de dados de treino
df_train = pd.read_csv('/content/drive/MyDrive/Projetos_DataScience/Ames_Iowa_Housing/train.csv')

#Ver as primeiras linhas da base de dados
print("--- Primeiras Linhas ---")
display(df_train.head())

#Ver as últimas linhas da base de dados
print("--- Últimas Linhas ---")
display(df_train.tail())

#Verificando o tamanho da base de dados
print("\n--- Tamanho da Base de Dados ---")
print(f"Linhas: {df_train.shape[0]} | Colunas: {df_train.shape[1]}")

#Verificando os tipos de dados
print("\n--- Tipos de Dados ---")
df_train.info()

#Verificando os valores ausentes
print("\n--- Valores Ausentes ---")
print(df_train.isnull().sum())
```

```python
# Calcular a quantidade de valores nulos por coluna
valores_nulos = df_train.isnull().sum()

# Filtrar apenas as colunas que possuem pelo menos 1 valor nulo
colunas_com_nulos = valores_nulos[valores_nulos > 0].sort_values(ascending=False)

# Calcular a porcentagem de valores nulos para ficar mais visual
porcentagem_nulos = (colunas_com_nulos / len(df_train)) * 100

# Juntar os dois dados em um DataFrame para analisar melhor
relatorio_nulos = pd.DataFrame({
    'Total de Nulos': colunas_com_nulos,
    'Porcentagem (%)': porcentagem_nulos
})

print("--- Colunas com Valores Ausentes no Treino ---")
display(relatorio_nulos)
```

```python
#Criando uma copia do banco de dados para preservar antes do tratamento
df_clean = df_train.copy()
```

```python
#Lista de colunas categoricas onde NaN significa "Nao tem" (none)
colunas_para_none = [
    'Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu',
    'Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond',
    'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2',
    'Mas Vnr Type'
]

#Substituindo NaN por "None"
for coluna in colunas_para_none:
    df_clean[coluna] = df_clean[coluna].fillna("None")
```

```python
#Lista de colunas numéricas onde NaN significa 0 ( area e quantidade )
colunas_para_zero = [
    'Garage Cars', 'Garage Area', 'Mas Vnr Area',
    'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF',
    'Bsmt Full Bath', 'Bsmt Half Bath'
]
#Substituindo NaN por 0 nessas colunas
for coluna in colunas_para_zero:
    df_clean[coluna] = df_clean[coluna].fillna(0)
```

```python
#Verificando como ficou apos tratamento
print("Tratamento concluido! Restam nulos nas colunas tratadas?")
print(df_clean[colunas_para_none + colunas_para_zero].isnull().sum().sum())
```

```python
# Mudar a configuração para exibir até 100 colunas sem omitir nenhuma
pd.set_option('display.max_columns', 100)
print(df_clean.head())
```

```python
#Verificando valores Nulos apos primeiro tratamento
valores_nulos = df_clean.isnull().sum()

# Filtrar apenas as colunas que possuem pelo menos 1 valor nulo
colunas_com_nulos = valores_nulos[valores_nulos > 0].sort_values(ascending=False)

# Calcular a porcentagem de valores nulos para ficar mais visual
porcentagem_nulos = (colunas_com_nulos / len(df_clean)) * 100

# Juntar os dois dados em um DataFrame para analisar melhor
relatorio_nulos = pd.DataFrame({
    'Total de Nulos': colunas_com_nulos,
    'Porcentagem (%)': porcentagem_nulos
})

print("--- Colunas com Valores Ausentes no Treino ---")
display(relatorio_nulos)
```

```python
# 1. Tratar 'Electrical': Preencher com a Moda (o valor mais comum)
moda_eletrico = df_clean['Electrical'].mode()[0]
df_clean['Electrical'] = df_clean['Electrical'].fillna(moda_eletrico)

# 2. Tratar 'Garage Yr Blt': Como a garagem não existe, vamos usar a mediana geral dos anos
mediana_garagem = df_clean['Garage Yr Blt'].median()
df_clean['Garage Yr Blt'] = df_clean['Garage Yr Blt'].fillna(mediana_garagem)

# 3. Tratar 'Lot Frontage': Preencher com a mediana da frente do terreno de cada bairro
df_clean['Lot Frontage'] = df_clean.groupby('Neighborhood')['Lot Frontage'].transform(lambda x: x.fillna(x.median()))

# --- VALIDAÇÃO FINAL ---
# Rodar novamente o verificador para cravar o ZERO absoluto no dataset inteiro
print("Quantidade total de nulos restantes no dataset inteiro:")
print(df_clean.isnull().sum().sum())
```

```python
# 1. Descobrir qual coluna ainda tem esses 2 nulos
nulos_restantes = df_clean.isnull().sum()
print("--- Coluna(s) culpada(s) ---")
print(nulos_restantes[nulos_restantes > 0])

# 2. Ver as linhas exatas onde isso aconteceu para entender o motivo
linhas_com_nulo = df_clean[df_clean.isnull().any(axis=1)]
print("\n--- Linhas afetadas ---")
display(linhas_com_nulo[['Neighborhood', 'Lot Frontage']])
```

```python
# Substituir os nulos restantes pela mediana global da coluna
df_clean['Lot Frontage'] = df_clean['Lot Frontage'].fillna(df_clean['Lot Frontage'].median())

# Salvar o arquivo final com o ZERO absoluto cravado
df_clean.to_csv('AmesHousing_train_clean.csv', index=False)

print("--- Verificação Final de Nulos no Dataset Inteiro ---")
print(f"Total de NaNs restantes: {df_clean.isnull().sum().sum()}")
```
